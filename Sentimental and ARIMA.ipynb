# Manufacturing Pred - ARIMA

```python
# Importing required libraries 
import numpy as np 
import pandas as pd 
#import matplotlib.pylot as plt 
from statsmodels.tsa.seasonal import seasonal_decompose 
  
# Read the AirPassengers dataset 
airline = pd.read_csv("GDP_growth(annual %).csv", 
                       index_col ='Year', 
                       parse_dates = True) 
  
# Print the first five rows of the dataset 
airline.head() 
  
# ETS Decomposition 
result = seasonal_decompose(airline['Manufacturing'],  
                            model ='additive') 
  
# ETS plot  
result.plot() 
```

```python

from pmdarima import auto_arima 

import warnings 
warnings.filterwarnings("ignore") 

stepwise_fit = auto_arima(airline['Manufacturing'], start_p = 1, start_q = 1, max_p = 3, max_q = 3, m = 12, start_P = 0, seasonal = True, d = None, D = 1, trace = True, error_action ='ignore',  suppress_warnings = True,stepwise = True)		 # set to stepwise 

stepwise_fit.summary() 
```

```python
train = airline.iloc[:len(airline)-20] 
test = airline.iloc[len(airline)-20:]  


from statsmodels.tsa.statespace.sarimax import SARIMAX 

model = SARIMAX(train["Manufacturing"], order = (0, 0, 1) ,seasonal_order=(0, 1, 1, 12)) 

result = model.fit() 
result.summary() 
```

```python
start = len(train) 
end = len(train) + len(test) - 1

# Predictions for one-year against the test set 
predictions = result.predict(start, end, typ = 'levels').rename("Predictions") 

# plot predictions and actual values 
predictions.plot(legend = True) 
test['Manufacturing'].plot(legend = True) 
```

```python
from sklearn.metrics import mean_squared_error 
from statsmodels.tools.eval_measures import rmse 

# Calculate root mean squared error 
print(rmse(test["Manufacturing"], predictions)) 

# Calculate mean squared error 
mean_squared_error(test["Manufacturing"], predictions) 
```

```python
# Train the model on the full dataset 
model = model = SARIMAX(airline['Manufacturing'], order = (0, 0, 1) ,seasonal_order=(0, 1, 1, 12)) 
result = model.fit() 

# Forecast for the next 10 years 
forecast = result.predict(start = len(airline), end = (len(airline)-1) + 10 , typ = 'levels').rename('Forecast') 

# Plot the forecast values 
airline['Manufacturing'].plot(figsize = (12, 5), legend = True) 
forecast.plot(legend = True) 
forecast
```
