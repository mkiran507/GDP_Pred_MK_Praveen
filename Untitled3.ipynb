```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=pd.read_csv("GDP_growth(annual %).csv")
#pd.set_option('display.float_format', lambda h: '%.5f' % h)
pd.reset_option('display.float_format')
dt=np.array(x)
data=dt[:,1:8]

ddt = pd.DataFrame(data)
ddt.columns = ['gdp','agri','manuf',
                     'industry','services','import','export']
kdt1 = ddt.fillna(ddt.mean())
ndt = kdt1/kdt1.max()


agric= ndt["agri"]
gdp=ndt["gdp"]
manuf=ndt["manuf"]
ind=ndt["industry"]
ser=ndt["services"]
imports=ndt["import"]
export=ndt["export"]

# ndt["export"].plot.line().set_xlabel('export', fontsize=14)
# ndt["gdp"].plot.line()
```

```python
print("Correlation of Agriculture on GDP:")
print(gdp.corr(agric))
print("\nCorrelation of Manufacturing on GDP:")
print(gdp.corr(manuf))
print("\nCorrelation of Industries on GDP:")
print(gdp.corr(ind))
print("\nCorrelation of Services on GDP:")
print(gdp.corr(ser))
print("\nCorrelation of Import on GDP:")
print(gdp.corr(imports))
print("\nCorrelation of Export on GDP:")
print(gdp.corr(export))

fig,a =  plt.subplots(2,3)
fig.set_size_inches(40,30)
a[0][0].scatter(agric,gdp, color='red')
a[0][0].set_title('GDP Vs Agriculture', fontsize=14)
a[0][0].set_xlabel('Agriculture', fontsize=14)
a[0][0].set_ylabel('GDP', fontsize=14)

a[0][1].scatter(manuf,gdp, color='red')
a[0][1].set_title('GDP Vs Manufacture', fontsize=14)
a[0][1].set_xlabel('Manufacture', fontsize=14)
a[0][1].set_ylabel('GDP', fontsize=14)

a[0][2].scatter(ind,gdp, color='red')
a[0][2].set_title('GDP Vs Industry', fontsize=14)
a[0][2].set_xlabel('Industry', fontsize=14)
a[0][2].set_ylabel('GDP', fontsize=14)

a[1][0].scatter(ser,gdp, color='red')
a[1][0].set_xlabel('Service', fontsize=14)
a[1][0].set_ylabel('GDP', fontsize=14)

a[1][1].scatter(imports,gdp, color='red')
a[1][1].set_xlabel('Import', fontsize=14)
a[1][1].set_ylabel('GDP', fontsize=14)

a[1][2].scatter(export,gdp, color='red')
a[1][2].set_xlabel('Export', fontsize=14)
a[1][2].set_ylabel('GDP', fontsize=14)
a[0][0].plot()
plt.show()

```

```python
y=ndt["gdp"]
X=ndt.drop("gdp",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=41)
X_train.shape
```

# Regression on Neural network

```python
from tensorflow import keras
regmodel = keras.models.Sequential()
regmodel.add(keras.layers.Dense(units=40,activation='relu', input_shape=(X_train.shape[1],)))
regmodel.add(keras.layers.Dense(units=40,activation='relu'))
regmodel.add(keras.layers.Dense(units=20, activation='relu'))
regmodel.add(keras.layers.Dense(units=20, activation='relu'))
regmodel.add(keras.layers.Dense(units=20, activation='relu'))
regmodel.add(keras.layers.Dense(units=10, activation='relu'))
regmodel.add(keras.layers.Dense(units=1,activation='linear'))

regmodel.compile(loss='mean_squared_logarithmic_error', optimizer='Adam', metrics=['mean_absolute_percentage_error'])
regmodel.summary()
```

```python
regmodel.fit(X_train, y_train, epochs = 1000, batch_size = 10,verbose=1, validation_split=0.2)
```

```python
regmodel.predict(X_test[:2])
```

```python
X_test[:2]
```

```python
y_test[:2]
```

```python
X_test.size
```

```python
regmodel.evaluate(X_test,y_test)
```

```python
ypred1=regmodel.predict(X_test[:])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test["agri"], y_test,  color='black')
plt.scatter(X_test["agri"], ypred1, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
from sklearn.neural_network import MLPRegressor   

model = MLPRegressor(solver='lbfgs',alpha=0.001,max_iter=1000).fit(X_train, y_train)
```

```python
model.predict(X_test[:2])
```

```python
model.score(X_test, y_test)
```

```python
ypred2=model.predict(X_test[:])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test["industry"], y_test,  color='black')
plt.scatter(X_test["industry"], ypred2, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
